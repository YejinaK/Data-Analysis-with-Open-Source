```python
import requests
import json

# 월 포맷 람다함수로 반복해서 만들기
format_m = lambda m: f"{m:02d}"
lambda_url = lambda y, m: f"http://openapi.seoul.go.kr:8088/6b6563797764707736335153575656/json/energyUseDataSummaryInfo/1/5/{y}/{format_m(m)}"
pairs = [(y, m) for y in range(2015, 2025) for m in range(1, 13)]

results = []

# 파라미터 호출
for y, m in pairs:
    data = None
    try:
        ## URL 및 파라미터 전송, raise_for_status<- 반응이 없을 때 처리하는 함수
        response = requests.get(lambda_url(y,m), timeout=5)
        response.raise_for_status()

        ## JSON 데이터 읽기
        data = response.json()

    except requests.exceptions.RequestException as e:
        print(f"API 호출 실패: {e}")
    except json.JSONDecodeError as e:
        print(f"JSON 파싱 실패: {e}")
    results.append((y, m, data))

if response.status_code == 200:
  print("api 호출 성공")
  print(response.json())
else:
  print(f"API 호출 실패: {response.status_code}")

```

```python
import pandas as pd

df = pd.read_json('output_df.json', encoding = "utf-8")
df.info()
```

```python
df["year"] = df["date"].dt.year
df
```

```python
months = df["date"].dt.month
def month_to_season(m:int) -> str:
    # 계절 규칙: 봄(3-5), 여름(6-8), 가을(9-11), 겨울(12-2)
    if 3 <= m <= 5:
        return "봄"
    elif 6 <= m <= 8:
        return "여름"
    elif 9 <= m <= 11:
        return "가을"
    else:
        return "겨울"

df["season"] = months.apply(month_to_season)

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 총 사용량 열 추가
df["total"] = df["electric"] + df["gas"] + df["water"] + df["district_heating"]

# 연도별 합계
yearly = df.groupby("year", as_index = False)["total"].sum()

# 그래프
plt.figure(figsize=(10, 5))
plt.plot(yearly["year"], yearly["total"])
plt.title("연도별 에너지 사용 총액 변화 - 3272")
plt.xlabel("연도")
plt.ylabel("총 사용량")
plt.grid(True)
plt.xticks(yearly["year"])
plt.ticklabel_format(style="plain", axis="y")
plt.savefig("연도별 에너지 사용 총액 변화 - 3272.png")
plt.show()
```

```python

# 2-2번 문제에서 만든 season 컬럼 그대로 사용 전제
import matplotlib.pyplot as plt
import seaborn as sns

# 계절별 평균 계산
season_avg = df.groupby("season")["gas"].mean()
order = ["봄", "여름", "가을", "겨울"]
season_avg = season_avg.reindex(order)

# 시각화
plt.figure(figsize=(7,5))
bars = plt.bar(season_avg.index, season_avg.values)
plt.title("계절별 가스 사용량 평균")
plt.xlabel("계절")
plt.ylabel("평균 가스 사용량")
plt.ticklabel_format(style="plain", axis="y")
plt.bar_label(bars, fmt="%.0f")
plt.show()
```
